# 1. Gráfico de Linhas

Tema: Pobreza no Brasil

Fonte: https://teoriaedebate.org.br/2018/03/07/com-o-aumento-da-extrema-pobreza-brasil-retrocede-dez-anos-em-dois/

```python
ano  = [2010,2011,2012,2013,2014,2015,2016]
extpobreza = [23000000, 21000000, 17000000, 16000000, 14000000, 17000000, 21000000]
pobreza = [8000000, 7500000, 6500000, 7000000, 5000000, 7000000, 10000000]

plt.plot(ano, extpobreza, label="Extrema Pobreza",color='g')
plt.plot(ano, pobreza, label="Pobreza",color='b')

plt.ylabel("Aproximação em Milhões")
plt.title("Pessoas em pobreza e extrema pobreza \n Brasil 2010 a 2016")
plt.legend()

#O código abaixo faz o seguinte:
#Impede que os valores das populações sejam escritos em notação científica.
plt.ticklabel_format(style='plain', axis='y') 
plt.grid(True)
#plt.box(False)
plt.show()
```

# 4. Gráfico de Dispersão

Tema: Escolarização no Brasil

Fonte: https://www.ibge.gov.br/cidades-e-estados

Sintaxe: https://www.youtube.com/watch?v=r238SpXA9As


```python
plt.title("Escolarização 6 a 14 anos (IBGE)")

ano = [2016, 2017, 2018, 2019]
percentual = [99.2, 99.2, 99.3, 99.7]

plt.xlim(2015, 2020)
plt.ylim(99,100)

plt.scatter(ano , percentual, linewidths = 5, marker='v', color='r')
plt.ticklabel_format(style='plain', axis='y')
plt.ylabel("% da População")
plt.grid(True)
plt.show()
```

# 5. Stack Plots

Tema: Pessoas com Deficiência no Brasil

Fonte: https://educa.ibge.gov.br/jovens/conheca-o-brasil/populacao/20551-pessoas-com-deficiencia.html

Sintaxe: https://matplotlib.org/stable/gallery/lines_bars_and_markers/stackplot_demo.html

```python
plt.title("Pessoas com Deficiência 2010 - IBGE")

plt.ylabel("% da População")

#Dados reais apenas em 2010
idxes = [2010,  2015,  2020]
arr1 = [3.4, 2, 4]
arr2 = [2.3, 2, 3]
arr3 = [1.1, 1, 2]
arr4 = [1.4, 1, 2]

plt.plot([], [], color='r', label = 'Visual')
plt.plot([], [], color='g', label = 'Motora')
plt.plot([], [], color='b', label = 'Auditiva')
plt.plot([], [], color='cyan', label = 'Mental/Intelectual')

plt.stackplot(idxes, arr1, arr2, arr3, arr4, colors= ['r', 'g', 'b','cyan'])

plt.legend()
plt.show()
```

# 6. Gráfico de Pizza

Tema: Quantidade de Homens e Mulheres no Brasil

Fonte: https://educa.ibge.gov.br/jovens/conheca-o-brasil/populacao/18320-quantidade-de-homens-e-mulheres.html

```python
plt.title("População residente, segundo o sexo (%) - IBGE 2019")

labels = "Homens","Mulheres"
colors = ["cyan","orange"]
sections = [48.2, 51.8]

plt.pie(sections, labels=labels, colors=colors,
        startangle=90, autopct = '%1.2f%%')

plt.show()
```

# 2) Faça uma pesquisa de um gráfico e procure repetir o mesmo criando através do Matplotlib abaixo:

Fonte: https://teoriaedebate.org.br/2018/03/07/com-o-aumento-da-extrema-pobreza-brasil-retrocede-dez-anos-em-dois/

```python
x1 = [1.6, 3.6, 5.6, 7.6, 9.6, 11.6]
y1 = [50, 39, 86, 72, 104, 53]
x2 = [2, 4, 6, 8, 10, 12]
y2 = [91, 67, 144, 189, 204, 93]

largura = 1.5
plt.bar(x1, y1, label="Pobreza", color='cyan', width= largura)
plt.bar(x2, y2, label="Extrema Pobreza", color='orange')
plt.plot()

plt.axis([0.5,12.7,0,250])
labels= ["Norte","Nordeste","Sudeste","Sul","Centro-Oeste","Brasil"]
plt.xticks(x1,labels=labels)

#plt.xlabel("Regiões")
plt.ylabel("Porcentagem %")
plt.title("Variação relativa dos quantitativos de pessoas em "+
          "Pobreza e Extrema Pobreza entre 2014 e 2016"+
          "\n Brasil e Grandes Regiões 2014-2016")
plt.legend()
plt.grid(True,color= 'black')
plt.show()
```
